```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
```

```python
page_to_scrape = requests.get('https://quotes.toscrape.com/')
#parse the website
soup = BeautifulSoup(page_to_scrape.text, 'html.parser')
quotes = soup.findAll("span", attrs= {"class":"text"})
authors = soup.findAll("small", attrs= {"class": "author"})
tags = soup.findAll("a", attrs= {"class":"tag"})
```

```python
file = open("scraped_quotes.csv", "w")
writer = csv.writer(file)
writer.writerow(["Quotes", "Authors", "Tags" ])
```

```python
#To get the quotes and authors using for loop
for quote, author, tag in zip(quotes, authors, tags):
  print(quote.text + " - " + author.text + " ," + tag.text)
  writer.writerow([quote.text, author.text, tag.text])
file.close()
```
